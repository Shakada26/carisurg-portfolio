The objective of this task was to clean the Temperature column from a clinical triage dataset. I converted the values into a numeric format to ensure consistency, checked for missing data, and removed unrealistic temperature values outside the normal human physiological range of 30 to 45 degrees Celsius. This was done to improve data quality and prepare the dataset for accurate analysis in later tasks.

```python
import pandas as pd

df = pd.read_csv('/content/EmergencyTriageDataset_Reduced_Dirty.csv')
display(df.head())
```

```python
# View raw temperature values
print("Before Cleaning:")
print(df["Temp"].head(10))

# Dataset shape before cleaning
print("\nDataset Shape Before Cleaning:")
print(df.shape)

# Convert to numeric
df["Temp"] = pd.to_numeric(df["Temp"], errors="coerce")

# Check for missing values
print("\nMissing Values:")
print(df["Temp"].isnull().sum())

# Remove impossible temperatures
df = df[(df["Temp"] >= 30) & (df["Temp"] <= 45)]

# Dataset shape after cleaning
print("\nDataset Shape After Cleaning:")
print(df.shape)

# Show cleaned values
print("\nAfter Cleaning:")
print(df["Temp"].head(10))

# Summary statistics
print("\nTemperature Summary:")
print(df["Temp"].describe())
```
