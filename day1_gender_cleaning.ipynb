Project Overview: This project establishes an automated data-cleaning pipeline for demographic entries within the Emergency Department Electronic Health Record (EHR) triage dataset at Mercer General Hospital. The primary objective is standardizing non-compliant string patterns in the Gender field to ensure rigorous clinical data safety for future downstream machine learning deployment.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/EmergencyTriageDataset_Reduced_Dirty.csv")

# Check Gender values before cleaning
print("Before Cleaning (raw):")
print(df["Gender"].unique())

# Clean Gender column to standardize string values
df["Gender"] = df["Gender"].astype(str).str.strip().str.upper()

# Standardize values, mapping '0' to 'Male' and '1' to 'Female' as strings first
df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "MALE": "Male",
    "0": "Male", # Map '0' to Male string
    "F": "Female",
    "FEMALE": "Female",
    "1": "Female" # Map '1' to Female string
})

# Encode 'Male' as 0 and 'Female' as 1
df["Gender"] = df["Gender"].replace({
    "Male": 0,
    "Female": 1
})

# Check Gender values after cleaning and encoding
print("\nAfter Cleaning and Encoding:")
print(df["Gender"].unique())

display(df)
```
